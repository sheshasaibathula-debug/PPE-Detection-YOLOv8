```python
import zipfile

zip_path = "/content/archive.zip"
extract_path = "/content/dataset"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Dataset Extracted Successfully!")
```

```python
import os

print(os.listdir("/content/dataset"))
```

```python
import yaml

data_yaml_content = {
    'train': '/content/dataset/train',
    'val': '/content/dataset/valid',
    'test': '/content/dataset/test',

    'nc': 2,
    'names': ['player', 'ball']
}

data_yaml_path = '/content/dataset/data.yaml'

with open(data_yaml_path, 'w') as file:
    yaml.dump(data_yaml_content, file, default_flow_style=False)

print("data.yaml created successfully!")
print(yaml.dump(data_yaml_content, default_flow_style=False))
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="/content/dataset/data.yaml",
    epochs=10,
    imgsz=640,
    batch=4
)
```
